# Daten

```python
qm_haus = 806
landesgrundsteuersatz = 0.00091 # Entspricht 0,91 Promille
hebesatz_wehingen = 3 # Entspricht 300%
```

```python
bodenrichtwert_2016 = [106]
for wert in range(400): # 106 + range(xxx), xxx = Bis Euro Wert
    bodenrichtwert_2016.append(bodenrichtwert_2016[wert]+1)

print(bodenrichtwert_2016)
```

```python
# betrag_ohne_steuer_liste = []
# betrag_mit_steuer_liste = []
betragssteuer_dict = {}

for wert in range(len(bodenrichtwert_2016)):
    # Bspw. 806 qm × 106 Euro/qm = 85.436 Euro
    betrag_ohne_steuer = qm_haus * bodenrichtwert_2016[wert]
    # betrag_ohne_steuer_liste.append(betrag_ohne_steuer)
    
    # Bspw. 85.436 Euro x 0,00091 % = 77,74676 €
    betrag_mit_steuer = betrag_ohne_steuer * landesgrundsteuersatz
    
    # Bspw. 77,74676 * 3 (entspricht 300%) = 233,24 €
    betrag_mit_steuer = betrag_mit_steuer * hebesatz_wehingen
    betrag_mit_steuer = round(betrag_mit_steuer, 2)
    # betrag_mit_steuer_liste.append(betrag_mit_steuer)
     
    
    betragssteuer_dict[bodenrichtwert_2016[wert]] = betrag_mit_steuer

# print(f"Betrag ohne Steuer:\n{betrag_ohne_steuer_liste}\n")

# print(f"Betrag mit Steuer:\n{betrag_mit_steuer_liste}\n")

print(f"Schema:\n Bodenrichtwert: Betrag mit Steuer und Hebel\n\n{betragssteuer_dict}\n")
```

# Visuelle Darstellung 

```python
import plotly.express as px

myList = betragssteuer_dict.items()
myList = sorted(myList) 
x, y = zip(*myList)     

px.scatter(myList, x=x, y=y,
           labels={
               "x":"Bodenrichtwert",
               "y":"Betrag (mit Steuermesszahl und Hebesatz)"
           },
           title="Neuer Grundsteuerbetrag")
    
```
