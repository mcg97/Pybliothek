## Data set cleaning

```python
games_data[games_data.isna().any(axis=1)] # Suche alle NaN Werte im DataFrame
```

```python
games_data_clean = games_data.copy()
games_data_clean = games_data_clean[~games_data_clean.isna().any(axis=1)]
games_data_clean
```

## Add home team name to dataframe

```python
# Merge two DataFrames (Home Team)

games_data_clean_team_name = games_data_clean.copy()
games_data_clean_team_name = games_data_clean_team_name.merge(right=teams_data[["TEAM_ID","NICKNAME"]],
                                                                        left_on="HOME_TEAM_ID",
                                                                        right_on="TEAM_ID",
                                                                        how="left")


games_data_clean_team_name["HOME_TEAM_ID"] = games_data_clean_team_name["HOME_TEAM_ID"].astype("object")
games_data_clean_team_name.rename(columns={"HOME_TEAM_ID" : "HOME_TEAM_NICKNAME"}, inplace=True)
games_data_clean_team_name["HOME_TEAM_NICKNAME"] = games_data_clean_team_name["NICKNAME"]
games_data_clean_team_name.drop(columns=["TEAM_ID", "NICKNAME"], inplace=True)

games_data_clean_team_name
```

## Add visitor team name to dataframe

```python
games_data_clean_team_name_final = games_data_clean_team_name.copy()
games_data_clean_team_name_final = games_data_clean_team_name_final.merge(right=teams_data[["TEAM_ID","NICKNAME"]],
                                                                        left_on="VISITOR_TEAM_ID",
                                                                        right_on="TEAM_ID",
                                                                        how="left")


games_data_clean_team_name_final["VISITOR_TEAM_ID"] = games_data_clean_team_name_final["VISITOR_TEAM_ID"].astype("object")
games_data_clean_team_name_final.rename(columns={"VISITOR_TEAM_ID" : "VISITOR_TEAM_NICKNAME"}, inplace=True)
games_data_clean_team_name_final["VISITOR_TEAM_NICKNAME"] = games_data_clean_team_name_final["NICKNAME"]
games_data_clean_team_name_final.drop(columns=["TEAM_ID", "NICKNAME"], inplace=True)

games_data_clean_team_name_final
```

```python
px.histogram(games_data_clean_team_name_final, x="HOME_TEAM_NICKNAME", color="SEASON").update_xaxes(categoryorder="total descending")
```

## Game Analysis

```python
# HOME_TEAM_WINS = 1 => Home team wins
# HOME_TEAM_WINS = 0 => Visitor team wins
```

### All Games 

```python
all_games_data = games_data_clean_team_name_final.copy()
all_games_data
```

#### Visitor team wins

```python
games_data_all_visitor = games_data_clean_team_name_final.copy()
games_data_all_visitor
```

```python
games_data_all_visitor_win = games_data_clean_team_name_final.copy()
games_data_all_visitor_win = games_data_all_visitor_win.loc[(games_data_all_visitor_win["HOME_TEAM_WINS"] == 0)]
games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] = games_data_all_visitor_win["PTS_away"] - games_data_all_visitor_win["PTS_home"]
games_data_all_visitor_win
```

```python
games_data_all_visitor_win.loc[(games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] >= 1) &
                               (games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] <= 9), "VISITOR_TEAM_PTS_DIFF_CLUSTER"] = "1-9"

games_data_all_visitor_win.loc[(games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] >= 10) &
                               (games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] <= 19), "VISITOR_TEAM_PTS_DIFF_CLUSTER"] = "10-19"

games_data_all_visitor_win.loc[(games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] >= 20) &
                               (games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] <= 29), "VISITOR_TEAM_PTS_DIFF_CLUSTER"] = "20-29"

games_data_all_visitor_win.loc[(games_data_all_visitor_win["VISITOR_TEAM_PTS_DIFF"] >= 30), "VISITOR_TEAM_PTS_DIFF_CLUSTER"] = "30+"

games_data_all_visitor_win
```

```python
px.histogram(games_data_all_visitor_win, x="VISITOR_TEAM_NICKNAME", color="VISITOR_TEAM_PTS_DIFF_CLUSTER").update_xaxes(categoryorder="total ascending", title_text="Visitor Team")
```

### Cleveland Cavaliers vs Charlotte Hornets

```python
games_data_cavs_hornets = games_data_clean_team_name_final.copy()
games_data_cavs_hornets
```

```python
games_data_cavs_hornets = games_data_cavs_hornets.loc[((games_data_cavs_hornets["HOME_TEAM_NICKNAME"] == "Hornets") |
                                                       (games_data_cavs_hornets["HOME_TEAM_NICKNAME"] == "Cavaliers")) &
                                                      ((games_data_cavs_hornets["VISITOR_TEAM_NICKNAME"] == "Hornets") |
                                                       (games_data_cavs_hornets["VISITOR_TEAM_NICKNAME"] == "Cavaliers"))]

games_data_cavs_hornets
```

#### Home win

##### Home points cluster

```python
home_cavs_hornets_pts_cluster_1 = games_data_cavs_hornets.copy()

home_cavs_hornets_pts_cluster_1.loc[(home_cavs_hornets_pts_cluster_1["PTS_home"] >= 70) & 
                                    (home_cavs_hornets_pts_cluster_1["PTS_home"] <= 80), "PTS_home_cluster"] = "70-80"
home_cavs_hornets_pts_cluster_1.loc[(home_cavs_hornets_pts_cluster_1["PTS_home"] >= 81) & 
                                    (home_cavs_hornets_pts_cluster_1["PTS_home"] <= 90), "PTS_home_cluster"] = "81-90"
home_cavs_hornets_pts_cluster_1.loc[(home_cavs_hornets_pts_cluster_1["PTS_home"] >= 91) & 
                                    (home_cavs_hornets_pts_cluster_1["PTS_home"] <= 99), "PTS_home_cluster"] = "91-99"
home_cavs_hornets_pts_cluster_1.loc[(home_cavs_hornets_pts_cluster_1["PTS_home"] >= 100), "PTS_home_cluster"] = "100+"

home_cavs_hornets_pts_cluster_1

```

```python
px.histogram(home_cavs_hornets_pts_cluster_1, 
             x="HOME_TEAM_NICKNAME", 
             color="PTS_home_cluster").update_xaxes(categoryorder="total ascending")
```

```python
px.histogram(home_cavs_hornets_pts_cluster_1, 
             x="HOME_TEAM_NICKNAME", 
             color="HOME_TEAM_WINS").update_xaxes(categoryorder="total ascending")
```

##### Point difference won

```python
home_cavs_hornets_pt_diff_won_2 = home_cavs_hornets_pts_cluster_1.copy()
home_cavs_hornets_pt_diff_won_2 = home_cavs_hornets_pt_diff_won_2.loc[home_cavs_hornets_pt_diff_won_2["HOME_TEAM_WINS"] == 1]
home_cavs_hornets_pt_diff_won_2["HOME_TEAM_PTS_DIFF"] = home_cavs_hornets_pt_diff_won_2["PTS_home"] - home_cavs_hornets_pt_diff_won_2["PTS_away"]
home_cavs_hornets_pt_diff_won_2
```

##### Points difference cluster

```python
home_cavs_hornets_pt_diff_won_pts_cluster_3 = home_cavs_hornets_pt_diff_won_2.copy()
home_cavs_hornets_pt_diff_won_pts_cluster_3.loc[(home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] >= 1.0) &
                                                (home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] <= 9.0), "HOME_TEAM_PTS_DIFF_CLUSTER"] = "1-9"

home_cavs_hornets_pt_diff_won_pts_cluster_3.loc[(home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] >= 10.0) &
                                                (home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] <= 19.0), "HOME_TEAM_PTS_DIFF_CLUSTER"] = "10-19"

home_cavs_hornets_pt_diff_won_pts_cluster_3.loc[(home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] >= 20.0) &
                                                (home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] <= 29.0), "HOME_TEAM_PTS_DIFF_CLUSTER"] = "20-29"

home_cavs_hornets_pt_diff_won_pts_cluster_3.loc[(home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"] >= 30.0), "HOME_TEAM_PTS_DIFF_CLUSTER"] = "30+"

home_cavs_hornets_pt_diff_won_pts_cluster_3
```

```python
px.histogram(home_cavs_hornets_pt_diff_won_pts_cluster_3, x="HOME_TEAM_NICKNAME", color="HOME_TEAM_PTS_DIFF_CLUSTER").update_xaxes(categoryorder="total ascending")
```

```python
home_cavs_hornets_pt_diff_won_pts_cluster_3["HOME_TEAM_PTS_DIFF"].mean()

# Average home team wins with +11,24 points
```
