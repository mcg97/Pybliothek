# Duration Analysis

```python
duration_analysis = data_import.copy()
duration_analysis["duration"].str.contains("min").value_counts() # 3 haben NaN werte
```

```python
duration_analysis["type"].value_counts()
```

```python
duration_analysis[duration_analysis["duration"].isnull()] 
```

```python
overwriter = duration_analysis.duration.isin([np.NaN])
duration_analysis.loc[overwriter, 'duration'] = duration_analysis.loc[overwriter, 'rating']
```

```python
duration_analysis["duration"].str.contains("min").value_counts() # 3 NaN werte beseitigt
```

```python
duration_analysis[duration_analysis["show_id"] == "s5542"]
# duration_analysis[duration_analysis["show_id"] == "s5795"]
# duration_analysis[duration_analysis["show_id"] == "s5814"]
```

```python
duration_timer = duration_analysis.copy()
duration_timer_split = duration_timer["duration"].str.split(" ", expand=True)

px.histogram(duration_timer_split, x=[0]).update_xaxes(categoryorder="total descending")

# 1, 2, 3 => Seasons => TV Series
```

```python
duration_timer = pd.concat([duration_timer, duration_timer["duration"].str.split(" ", expand=True)], axis=1)
duration_timer
```

```python
duration_timer_min = duration_timer[duration_timer[1] == "min"] # min => Filme
duration_timer_min["duration_movie"] = duration_timer_min[0]
duration_timer_min = duration_timer_min.drop(columns=0)
duration_timer_min = duration_timer_min.drop(columns=1)
duration_timer_min
```

```python
duration_timer_min["duration_movie"] = duration_timer_min["duration_movie"].astype("int64") # Umwandeln in int
duration_timer_min.dtypes
```

```python
# Cluster machen (Kurz, mittellang, lang, ...)

duration_timer_min.loc[(duration_timer_min['duration_movie'] <= 30), 'duration_movie_cluster'] = 'kurz'
duration_timer_min.loc[(duration_timer_min['duration_movie'] >= 30) & (duration_timer_min['duration_movie'] <= 60), 'duration_movie_cluster'] = 'mittelkurz'
duration_timer_min.loc[(duration_timer_min['duration_movie'] >= 60) & (duration_timer_min['duration_movie'] <= 120), 'duration_movie_cluster'] = 'lang'
duration_timer_min.loc[(duration_timer_min['duration_movie'] >= 120), 'duration_movie_cluster'] = 'ueberllang'

duration_timer_min 
```

```python
px.histogram(duration_timer_min, x="rating", color="duration_movie_cluster")
```

# Director Analysis

```python
director_analysis = data_import.copy()
director_analysis = pd.concat([director_analysis, data_import["director"].str.split(",", expand=True)], axis=1)
director_analysis_clean = director_analysis[director_analysis["director"].notna()]
director_analysis_clean
```

```python
director_analysis_clean = director_analysis_clean.melt(id_vars=["type","title"], value_name="Director", value_vars=range(12)) # Wegen 0-11 oben 
director_analysis_clean = director_analysis_clean[director_analysis_clean["Director"].notna()] # NaN Werte löschen
director_analysis_clean
```
