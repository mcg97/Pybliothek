## Reihen mathematische Anwendungen 

```python
df_2_clean = pd.read_csv("data/pandas/5_ReihenSauebern_salary_df.csv")
df_2_clean
```

```python
# Maximum

max_df_2_clean = df_2_clean["salary_clean"].cummax()
print(max_df_2_clean,"\n\n")
max_df_2_clean_value = max_df_2_clean[(len(max_df_2_clean)-1)] 
print(f"Maximaler Wert im Salary: {max_df_2_clean_value} $")
```

```python
# Minimum
 
min_df_2_clean = df_2_clean["salary_clean"].cummin()
print(min_df_2_clean,"\n\n")
min_df_2_clean_value = min_df_2_clean[(len(min_df_2_clean)-1)] 
print(f"Minimaler Wert im Salary: {min_df_2_clean_value} $")
```

```python
# Maximaler Produktwert bevor es zu "inf" wird
prod_df_2_clean = df_2_clean["salary_clean"].cumprod()
print(prod_df_2_clean,"\n\n")
rows_length = len(prod_df_2_clean)
my_dict = {}
for x in range(0, rows_length):
    if(last_prod_num != np.inf):
        print("1: ",last_prod_num)
        my_dict[x] = last_prod_num
        last_prod_num = prod_df_2_clean[x]
        print("2: ",last_prod_num)
    else:
        last_prod_num = 0
        print("Beendet")
        break
print("\n")
max_value = max(my_dict.values())
max_key = max(my_dict, key=my_dict.get)
print(f"Key: {max_key} Wert: {max_value}")
```

## Handling NA

```python
dict_4 = {"A": [1,2,np.nan,9,np.nan],
          "B": [123,np.nan, np.nan,22,np.nan],
          "C": [7., 8., 9., 19., 29.]}
df_nan = pd.DataFrame(dict_4)

print(df_nan,"\n")

print("Zeilenweise:\n",df_nan.dropna(axis=0),"\n") 
print("Spaltenweise:\n",df_nan.dropna(axis=1),"\n") 
```

```python
print("Zwei pro Zeile:\n",df_nan.dropna(thresh=2, axis=0),"\n")
print("Behalte Spalten mit über 80 Prozent validen Daten:\n",
      df_nan.dropna(how="any",axis=0, thresh=df_nan.shape[1]*0.8),"\n")
```

```python
print("Mit mean von A Werten füllen:\n", 
      df_nan.fillna(value=df_nan["A"].mean()),"\n")

print("Mit vorherigem Wert NaN ersetzen, bis neuer Wert eintritt :\n", 
      df_nan.fillna(method="ffill"))
```

## Data Cleaning

```python
data_import = pd.read_csv("data/pandas/data_cleaning_challenge.csv")
data_import
```

```python
# Überflüssige Spalten löschen
data_drop = data_import.drop(columns=["Unnamed: 9", "Unnamed: 10"])
data_drop
```

```python
# NaN Werte aus der Spalte "Row Type" nehmen, for-Schleife Fehlermeldung (TypeError) vermeiden
no_nan = data_drop[data_drop["Row Type"].notna()]
no_nan
```

```python
# Liste für die Zuordnung
column_values = []
iteration = 0

for i in no_nan["Row Type"]:
    if "first name" in i:
        iteration += 1
    column_values.append(iteration)
# Zeilen_ID:0-11 Wert 1, 
# Zeilen_ID:14-25 Wert 2,
# ...
```

```python
column_values
```

```python
# Neue Spalte hinzufügen
iter_column_added = no_nan
iter_column_added["Iteration"] = column_values
iter_column_added
```

```python
# Die Einträge Reihe mit "Row Type" löschen (siehe Zelle oben Zeile 2) 
drop_extra_column_rows = iter_column_added[iter_column_added["Row Type"] != "Row Type"]
drop_extra_column_rows
```

```python
# Alle Namen in in DataFrame speichern
name_dataframe = drop_extra_column_rows[drop_extra_column_rows["Row Type"].str.contains("first name")]
name_dataframe
```

```python
name_dataframe = name_dataframe.loc[:,["Row Type", "Iter Number", "Power1", "Iteration"]]
name_dataframe
```

```python
name_dataframe = name_dataframe.loc[:,["Row Type", "Iter Number", "Power1", "Iteration"]]
name_dataframe.rename(columns={"Row Type": "First Name", "Iter Number":"Last Name", "Power1": "Date"}, inplace=True)# Änderung wird sofort übernommen, ansonsten wird eine Kopie des DataFrames zurückgegeben
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:] # First Name etc. aus den Zeilen entfernen
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[10:]
name_dataframe["Date"] = name_dataframe["Date"].str[5:]
name_dataframe
```

```python
no_name_dataframe = drop_extra_column_rows[~drop_extra_column_rows["Row Type"].str.contains("first name")]
no_name_dataframe
```

```python
no_name_dataframe_merged = pd.merge(left=name_dataframe, right=no_name_dataframe, how="inner", on="Iteration")
no_name_dataframe_merged
```

## Funktionen und Daten

```python
cs_df = pd.read_csv("data/pandas/ComputerSales.csv")

# pd.options.display.max_rows=10

def get_profit_total(df):
    prof_ser = df["Profit"]
    print(f"Total Profit: {prof_ser.sum()}")
get_profit_total(cs_df)
cs_df

def split_name(df):
    def get_names(full_name):
        f_name, l_name = full_name.split()
        return pd.Series(
            (f_name, l_name),
            index=["First Name", "Last Name"]
        )
    names = df["Contact"].apply(get_names)
    df[names.columns] = names
    return df

split_name(cs_df).head()

def create_age_groups(df):
    bins = [0, 30, 50, sys.maxsize]
    labels = ["<30", "30-50", ">50"]
    age_group = pd.cut(df["Age"],bins=bins, labels=labels)
    df["Age Group"] = age_group
    return df

create_age_groups(cs_df)
```
